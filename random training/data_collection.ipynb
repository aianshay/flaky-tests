```python
import pandas as pd
import os
```

```python
dataset = pd.DataFrame()

tokens = []
project_names = []
flakiness = []

projects = os.listdir('msr/flaky')
for project in projects:
    project_path = f'msr/flaky/{project}'
    files_per_project = os.listdir(project_path)
    
    for file_name in files_per_project:
        with open(project_path + '/' +  file_name, 'r') as reader:
            tokens.append(reader.read().replace("\n"," "))
            project_names.append(project)
    

dataset['token'] = tokens
dataset['project'] = project_names
dataset['flaky'] = 1
```

```python
dataset
```

```python
dataset2 = pd.DataFrame()
tokens = []
project_names = []
flakiness = []

projects = os.listdir('msr/non_flaky')
for project in projects:
    project_path = f'msr/non_flaky/{project}'
    files_per_project = os.listdir(project_path)
    
    for file_name in files_per_project:
        with open(project_path + '/' +  file_name, 'r', encoding='ANSI') as reader:
            tokens.append(reader.read().replace("\n"," "))
            project_names.append(project)
    
dataset2['token'] = tokens
dataset2['project'] = project_names
dataset2['flaky'] = 0
```

```python
dataset2
```

```python
final = dataset.append(dataset2)
```

```python
data = final.sample(frac=1).reset_index(drop=True)
```

```python
data.to_csv('msr/msr_data.csv', index=False)
```
